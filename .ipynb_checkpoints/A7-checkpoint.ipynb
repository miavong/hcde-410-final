#### Code - Data Preparation
The following code imports the necessary python programs to create graphs, plots, and various data visualizations. The code loads the csv files my A7 notebook and makes sure it is encoded correctly. It renames the country columns in both datasets so that we can merge the countries that exist in both datasets. I then create a master dataset between the two csv files (AI Global Impact and The Economic Freedom Index) for correctly merging countries.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets using the correct file paths
economic_freedom_data = pd.read_csv('economic_freedom_index2019_data_utf8.csv', encoding='ISO-8859-1')
ai_index_data = pd.read_csv('AI_index_db.csv', encoding='ISO-8859-1')

# Rename 'Country' column to 'country' for consistency
economic_freedom_data.rename(columns={"Country": "country"}, inplace=True)
ai_index_data.rename(columns={"Country": "country"}, inplace=True)

# Merge the datasets to keep only common countries
merged_data = pd.merge(economic_freedom_data, ai_index_data, on='country', how='inner')

# Checking how many countries are in the dataset
# num_countries = len(merged_data['country'].unique())
# print(num_countries)
```

# Analysis

## RQ 1 - AI Investment & Economic Freedom Index 

### Methodology 

The first research question we are exploring is: **Does a country's investment in AI correlate with its economic freedom?** To answer this question, we need to understand how two of the most broad key metrics, the Economic Freedom Index score and the AI Global Index score, are related. For context, the Economic Freedom Index is defined by Heritage.org as, "has measured economic freedom’s advances and retreats around the globe country by country, correlating those changes with fundamental measures of economic well-being like economic growth, reductions in poverty, various social indicators like longevity and health, and environmental protection." The AI Global Index is defined within the README of the dataset as an, "index to benchmark nations on their level of investment, innovation, and implementation of artificial intelligence." Thus, the Economic Freedom Score provides an indication of a country's level of economic prosperity and freedom (from 2019), while the AI Global Score reflects a country’s investment and capability in AI technologies. To see how the two scores are correlated, I will produce a scatter plot to help us visually explore the relationship between economic freedom and AI investment across countries. I've also added a regression line to represent the rate of change of the AI Index Score with respect to changes in the Economic Freedom Score. This will help us visualize the magnitude of a positive, negative, or no correlation between the variables. A positive slope would indicate that an increase in the Economic Freedom Score is generally associated with an increase in AI investment, a negative slope would indicate the opposite, and the magnitude of the slope determines the strength of this relationship.

I have also included a table that correlates the countries in our merged dataset with their AI Global Index and Economic Freedom scores for reference to what countries are represented in the dataset. This table will be used as a reference for later evaluation as we look into employment rates and AI work talent in the next research question.


#### Code - Research Question 1 - Economic Freedom vs AI Global Index Score Scatterplot

This code imports a linear regression package and uses our merged dataset to plot the countries based on their AI investment and Economic Freedom scores. It then produces a regression model by calculating the p-value from the linear slope. I then initialize all of this by labeling my axes for creating a scatter plot, each dot on the scatter plot represents a country.


```python
from scipy.stats import linregress

# Plot 'Economic Freedom Score' (from dataset 1) vs 'Total score' (from dataset 2)
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['2019 Score'], merged_data['Total score'], label='Country')

# Calculate and plot linear regression line (slope)
slope, intercept, r_value, p_value, std_err = linregress(merged_data['2019 Score'], merged_data['Total score'])
regression_line = slope * merged_data['2019 Score'] + intercept
plt.plot(merged_data['2019 Score'], regression_line, color='red', label=f'Regression Line (slope = {slope:.2f})')

# Labeling axes and adding title
plt.xlabel('Economic Freedom Score')
plt.ylabel('AI Index Total Score')
plt.title('Economic Freedom vs AI Investment Score')
plt.grid(True)
plt.legend()
plt.show()

```

#### Code - AI Global Index & Economic Freedom Score Visual Table 
This code creates a 3 column visual table using the same merged dataset from above. It initializes some sizing and formatting requirments so that the data fits evenly. It also displays the countries with the highest AI global index score in decending order from top to bottom with it's respective economic freedom score. 

```python
# Create a visual table of the countries and their scores
# Selecting the two columns we need: 'country', '2019 Score', 'Total score'
merged_data_sorted = merged_data[['country', '2019 Score', 'Total score']]
merged_data_sorted.columns = ['Country', 'Economic Freedom Score', 'AI Investment Score']
merged_data_sorted = merged_data_sorted.sort_values(by='AI Investment Score', ascending=False)

# Set up the plot for the table with a adaptive layout 
fig, ax = plt.subplots(figsize=(10, len(merged_data_sorted) * 0.4)) 

# Hide axes for table plot
ax.axis('tight')
ax.axis('off')

# Create the table
table = ax.table(cellText=merged_data_sorted.values, colLabels=merged_data_sorted.columns, cellLoc='center', loc='center')

# Adjust the size of the table 
table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width(col=list(range(len(merged_data_sorted.columns))))

# Display the table
plt.show()
```

### Findings 

The scatter plot visualization indicates that there may not be a strong correlation between the economic freedom score and AI investment for the countries analyzed. While there is a slight positive correlation, the calculated regression line indicates the correlation is not very strong. A slope of 0.73 indicates a positive relationship between the Economic Freedom Score and AI investment, which means that countries that have higher economic prosperity and freedom are generally associated with higher investments in AI technologies. However, it's a relatively small slope (less than 1.0), which suggests that while there is some positive correlation between economic freedom and AI investment, there may not be a strong direct correlation between these variables. Thus, we can conclude from this data analysis that countries that are economically more free are most likely investing more in AI, but we cannot confidently correlate the two. This conclusion suggests that other specific factors/variables may also determine AI investment and economic freedom, as we will explore later in this report.



### Limitations 

One limitation to this analysis was the presence of a few outliers, notably a country with a very high AI Index Total Score and Economic Freedom Score. In addition, a country with a high AI Index Score and a lower Economic Freedom Score could skew the overall trend and strength of our correlation. There could also be confounding variables such as government policies, cultural aspects, or political conditions that aren't accounted for in this analysis. Lastly, the Economic Freedom dataset was calculated in 2019, while the AI dataset was calculated in 2023. This time gap means that the economic freedom and AI investment scores could have changed significantly between these years, which limits our ability to accurately determine if there is a current relationship between economic freedom and AI investment in 2024.


## RQ 2 - Unemployment & AI Talent 

### Methodology 

The second research question we are exploring is: **How have national unemployment rates influenced efforts to create AI-related jobs and develop AI-driven solutions?** I had to alter this research question from my original project proposal, "How does a countries efforts into creating AI related jobs factor into the countries unemployment percentage?" because the Economic freedom dataset is from 2019 and the AI Index is from 2022 thus a question asking about how recent AI jobs influence past unemployment wouldn't make logical sense. Instead, this question is asking specifically if a country's rate of unemployment and job availability has influenced the amount of AI-related jobs currently on the market. I want to look directly at job security because in recent years the general public has feared how AI could potentially steal jobs away from working people. I've decided to create two bar graphs that compare the countries on unemployment rate and AI talent from the two datasets. To define these metrics, a country's unemployment rate (%) is, "a measure of the portion of the workforce that is not employed but is actively seeking work. Data are from International Labour Organization, World Employment and Social Outlook: Trends 2019." [https://www.heritage.org/index/pages/about#indexMethodology]. A country's AI talent is a numerical, "indicator (that) focuses on the availability of skilled practitioners for the provision of artificial intelligence solutions" [AI Index READ.ME]. These bar plots will help us visualize how unemployment rates might have affected a country's ability to develop AI talent. I chose not to create a stacked or double bar chart because the range of scores and percentages between the two datasets would be easily visualized, thus I had to make two separate bar plots.


#### Code - Unemployment Rates vs AI Talent Bar Graphs 

I created a copy of our merged dataset looking at the specific unemployment percentage and talent columns in the two datasets. I then made sure to get rid of any countries that were missing data from the new merged dataset. I then created bar plots for each variable, talent (AI index dataset) and Unemployment Rate (Economic freedom dataset) for each country and defined their axes. I also made the countries ascend from lowest scores/percentages to highest for both graphs for easier cross-examination visibility.


```python
# Create a copy of our merged dataset with the specific columns 
merged_data_filtered = merged_data[['country', 'Unemployment (%)', 'Talent']].copy()

# Drop any rows with missing values
merged_data_filtered.dropna(inplace=True)

# Sort countries by AI Talent Score for better visualization
merged_data_sorted_talent = merged_data_filtered.sort_values(by='Talent', ascending=True)

# Create a bar plot for AI Talent Score for each country
plt.figure(figsize=(20, 10))
plt.bar(merged_data_sorted_talent['country'], merged_data_sorted_talent['Talent'], color='orange')
plt.xlabel('Country', fontsize=14)
plt.ylabel('AI Talent Score', fontsize=14)
plt.title('AI Talent Score by Country', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()

# Sort countries by Unemployment Rate 
merged_data_filtered['Unemployment (%)'] = pd.to_numeric(merged_data_filtered['Unemployment (%)'], errors='coerce')
merged_data_sorted_unemployment = merged_data_filtered.sort_values(by='Unemployment (%)', ascending=True)

# Create a bar plot for Unemployment Rate for each country
plt.figure(figsize=(20, 10))
plt.bar(merged_data_sorted_unemployment['country'], merged_data_sorted_unemployment['Unemployment (%)'], color='blue')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Unemployment Rate (%)', fontsize=14)
plt.title('Unemployment Rate (%) by Country', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()

# I used these lines of code to fact check the bar plot tables 
#print(merged_data_sorted_talent)
#print(merged_data_sorted_unemployment)
```

### Findings 
The top contributors to AI talent in this subset are the US, India, and the UK. This roughly correlates with the top AI Global Index countries as well, although China is not in the top 3 countries ranked for AI talent scores. Overarchingly, we see that these three countries fall within the middle to low rank for unemployment percentages. These 3 countries have high AI Talent Scores and a relatively low/mid Unemployment Rate possibly because they have more resources and stability to invest in AI training and education, which would lead to developing higher AI talent. Most notably is India, ranking 2nd highest in AI talent and 12th lowest in unemployment (fairly low unemployment %). This particularly interests me because we've learned how data crowdsourcing work is often allocated to workers in India (as seen in analyzing the Amazon Mturk worker forums). I wish the AI index dataset had a breakdown of the type of AI talent per country to further investigate this finding. In addition, countries with higher unemployment rates such as South Africa, Greece, and Armenia have lower AI talent scores, which could indicate that the impact of AI on the overall workforce could be limited, either because AI hasn't been adopted enough to affect job markets significantly or because these countries lack the necessary skilled professionals to implement AI. In conclusion, the countries that are currently leading with high AI talent and job work had relatively low/mid unemployment rates in 2019, which could be a factor in their ability to invest in AI today.


### Limitations 

Because we merged the two datasets with only countries that have data from both the AI Index and Economic Freedom datasets, it's important that we don't generalize these rankings from a global point of view because this is only a limited sample size. For example, in our AI talent graph, France has the 8th highest AI talent; while this may be true for our subset of data, there could be higher talent-ranked countries that are not represented because they lacked unemployment % data or Economic Freedom Index data. Secondly, the unemployment rate data is from 2019, while the AI talent data was more recently collected in 2022. This time gap might lead to our AI talent and unemployment comparisons not aligning perfectly, as the unemployment rate from 2019 might not accurately represent the economic conditions at the time of AI investment reported. Lastly, economic conditions can change rapidly for every individual country, resulting in mass job insecurity and recessions - these confounding conditions are not factored into our methodology, which could lead to oversimplified conclusions.


## RQ 3 - Intellectual Property & AI Research 

### Methodology 

The third and final research question is: **Does a government's respect for intellectual property influence their efforts ing AI research?** The reasoning behind this research question stems from the idea that a government that respects and enforces intellectual property rights may be more likely to invest in AI research, resulting in an overall stronger investment in AI. As defined on heritage.org from the Economic Freedom dataset, the Property Rights measurement is, "...assesses the extent to which a country’s legal framework allows individuals to acquire, hold, and utilize private property and the extent to which these rights are secured by applicable laws that the government enforces effectively." [https://www.heritage.org/index/pages/about#indexMethodology]. A component of this measurement is the respect for intellectual property rights, which is how we will be interpreting the Property Rights value for the purpose of this research question. The value "Research" from the AI Index dataset is an indicator that "...focuses on the extent of specialist research and researchers; investigating the amount of publications and citations in credible academic journals." Intellectual property laws provide incentives for researchers and companies to innovate by ensuring that their inventions and discoveries are protected from unauthorized use or reproduction. Thus, a strong intellectual property environment could lead to greater government and private sector efforts to fund AI research, which would result in more specialized AI research work and publications.



### Code - Property Rights & AI Research vs Countries

The provided code analyzes the relationship between a government's respect for intellectual property and its investment in AI research. I created a copy of our merged dataset (containing the "Property Rights" score) and the AI Global Impact dataset (containing the "Research" score). I then filtered and cleaned the data by dropping rows with missing values. The code then creates bar plots to visualize the "Property Rights" scores and "AI Researcg" for each country. To explore the relationship between these two variables, a scatter plot is generated, with each point representing a country. Additionally, a linear regression line is added to the scatter plot to visually assess whether there is a positive or negative correlation between the respect for intellectual property andAI research.


```python
# Create a copy 
merged_data_filtered = merged_data[['country', 'Property Rights', 'Research']].copy()

# Drop any rows with missing values
merged_data_filtered.dropna(inplace=True)

# Sort countries by Property Rights Score for better visualization
merged_data_sorted_property = merged_data_filtered.sort_values(by='Property Rights', ascending=False)

# Create a bar plot for Property Rights Score for each country
plt.figure(figsize=(14, 8))
plt.bar(merged_data_sorted_property['country'], merged_data_sorted_property['Property Rights'], color='green')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Property Rights Score', fontsize=14)
plt.title('Property Rights Score by Country', fontsize=16)
plt.xticks(rotation=90, fontsize=10)
plt.tight_layout()
plt.show()

# Sort countries by AI Research for better visualization
merged_data_sorted_research = merged_data_filtered.sort_values(by='Research', ascending=False)

# Create a bar plot for AI Research for each country
plt.figure(figsize=(14, 8))
plt.bar(merged_data_sorted_research['country'], merged_data_sorted_research['Research'], color='purple')
plt.xlabel('Country', fontsize=14)
plt.ylabel('AI Research', fontsize=14)
plt.title('AI Research by Country', fontsize=16)
plt.xticks(rotation=90, fontsize=10)
plt.tight_layout()
plt.show()

# Linear regression line to understand the relationship between Property Rights and AI Research
from scipy.stats import linregress

slope, intercept, r_value, p_value, std_err = linregress(merged_data_filtered['Property Rights'], merged_data_filtered['Research'])
regression_line = slope * merged_data_filtered['Property Rights'] + intercept

plt.figure(figsize=(10, 6))
plt.scatter(merged_data_filtered['Property Rights'], merged_data_filtered['Research'], label='Data Points')
plt.plot(merged_data_filtered['Property Rights'], regression_line, color='red', label=f'Regression Line (slope = {slope:.2f})')
plt.xlabel('Property Rights Score', fontsize=14)
plt.ylabel('AI Research', fontsize=14)
plt.title('Property Rights vs AI Research with Regression Line', fontsize=16)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### ***Disclaimer: ChatGPT was used to help format syntax, debug and produce data visualization in the code above***

### Findings
The bar plot visualizations show that countries like Singapore, New Zealand, and Hong Kong rank high in respect for property rights, while the United States and China lead in AI research. From the scatter plot and regression line, we observe a positive slope (0.50), this suggests that countries with stronger intellectual property protection are somewhat more likely to invest in AI research. However, this correlation is not very strong, indicating that intellectual property rights are not the only factor driving increased AI research. An additional interesting note is that the scatterplot produced in this analysis closely resembles the AI Global Index & Economic Freedom graph from RQ1. We also notice that the two outlier countries in both graphs are the US and China. This could suggest that the US and China are currently the biggest global AI leaders in many AI developmental fields and may share similar economic and AI investment tendencies. While the correlation between intellectual property rights and AI research is largely inconclusive, one intriguing observation outside of is India's relatively low ranking in AI research despite having the second-highest score for AI talent. This could indicate that the type of AI work performed in India is less focused on academia and research and more centered on computing or development work.


### Limitations
The relationship between property rights and AI research may be influenced by other confounding variables, such as a country's overall level of education, college tuition support, or specific government incentives, which were not included in this study. In addition, for the purpose of this research question, we made the correlation of the Property Right score to represent a government's respect for intellectual property, but in reality, there are other factors to the property right score other than intellectual property. In addition, the AI index dataset doesn't specify how they define the qualifications of "specialist research and researchers" that are performing AI research. This may result in data being collected inconsistently or inaccurately and could compromise the quality of the dataset and the validity of the analysis as we interpret it within the research question.
